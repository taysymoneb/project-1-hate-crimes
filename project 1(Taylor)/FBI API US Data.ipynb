3.What are the trends for hate crime incident reportings in the US and NC from 2016-2022?

Between 2016 and 2022, there has been a definite increase in hate crime occurrences both nationally and locally in North Carolina. However, the number of incidents reported in North Carolina decreased by 15% from 2018 to 2019. After that, there was a drastic 65% increase in hate crimes reported. Nationally, there was a 9% increase in hate crime incidents from 2018 to 2019, followed by a 26% increase between 2019 and 2020.

```python
Between 2016 and 2022, there has been a definite increase in hate crime occurrences both nationally and locally in North Carolina. However, the number of incidents reported in North Carolina decreased by 15% from 2018 to 2019. After that, there was a drastic 65% increase in hate crimes reported. Nationally, there was a 9% increase in hate crime incidents from 2018 to 2019, followed by a 26% increase between 2019 and 2020.
# Dependencies and Setup
import requests
import pandas as pd
import json
from pprint import pprint
import matplotlib.pyplot as plt

#import api key from py file 
from api_key import fbi_api_key
```

```python
#build url
url = f"https://api.usa.gov/crime/fbi/cde/hate-crime/national/count/all?from=1991&to=2020&API_KEY={fbi_api_key}"
#jsonify response
response = requests.get(url).json()
response
```

```python
#create empty list to hold information to create a new data frame
years = []
incident_counts = []

#loop through the data to get information
for num in range(len(response["data"])):
    #print(response["data"][num])
    years.append(response["data"][num]["data_year"])
    incident_counts.append(response["data"][num]["incident_count"])
    
#create dataframe
incidentDF = pd.DataFrame({"Year":years, "Number of Incidents": incident_counts})
incidentDF = incidentDF.sort_values(by=["Year"])
incidentDF = incidentDF.reset_index(drop=True)
incidentDF
```

```python
#create a line graph displaying the years and incident counts
x = ["2016", "2017","2018", "2019", "2020", "2021", "2022"]
y = [6270, 7329, 7180, 7875, 9952, 10891, 11643]


plt.plot(x,y, marker="*", mfc = "lightblue", mec= "black", mew= "0.7", color= "black", markersize= 10)
plt.title ("Hate Crime Trend Nationally (2016-2022)")
plt.xlabel("Years")
plt.ylabel("Number of Incidents")
plt.savefig("output_data/Fig10.png")
plt.show()
```
