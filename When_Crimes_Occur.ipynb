```python
# Import Dependencies
import pandas as pd
from datetime import datetime
import datetime as dt
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
# Set path for file
crime_path = Path("Resources/hate_crime.csv")
```

```python
# Read the modified Hate Crimes csv and store into Pandas Dataframe
crime_df = pd.read_csv(crime_path, encoding="utf-8")
crime_df
```

```python
# Show specific columns of dataset
crime_df[["incident_id", "incident_date", "state_name", "total_offender_count", "victim_count", "offense_name", "location_name"]]
```

```python
crime_df["incident_date"] = pd.to_datetime(crime_df["incident_date"])
```

```python
crime_df.dtypes
```

```python
incident_date = crime_df["incident_date"]
incident_date = incident_date.dt.month_name()
incident_date
```

```python
crime_df["month"] = incident_date
crime_df.head()
```

```python
crime_df["month"].unique()
```

```python
crime_df["month"].value_counts()
```

```python
crime_counts = crime_df["month"].value_counts()
crime_counts = crime_counts.reindex(index = ["January", "February", "March", "April", "May", "June", "July",
                                            "August", "September", "October", "November", "December"])
crime_counts.plot(kind = "bar", title = "No. of Hate Crimes") 
plt.savefig("output_graphics/Fig1.png")
```
